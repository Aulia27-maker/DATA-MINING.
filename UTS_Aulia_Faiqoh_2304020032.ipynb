**DETEKSI OUTLIER**

````python
def cek_outlier_iqr(data, columns):
    hasil_outlier = []

    for col in columns:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        iqr = q3 - q1

        batas_bawah = q1 - 1.5 * iqr
        batas_atas = q3 + 1.5 * iqr

        jumlah_outlier = data[
            (data[col] < batas_bawah) | (data[col] > batas_atas)
        ].shape[0]

        hasil_outlier.append({
            'Fitur': col,
            'Jumlah_Outlier': jumlah_outlier,
            'Batas_Bawah': batas_bawah,
            'Batas_Atas': batas_atas
        })

    return pd.DataFrame(hasil_outlier)
````

````python
outlier_summary = cek_outlier_iqr(data_train, feature_columns)
outlier_summary
````

Kode tersebut digunakan untuk mendeteksi outlier pada setiap fitur numerik menggunakan metode IQR atau Interquartile Range. Metode ini menghitung nilai kuartil pertama (Q1), kuartil ketiga (Q3), dan selisihnya yang disebut IQR.

Batas outlier ditentukan dengan rumus:

Batas bawah = Q1 - 1.5 × IQR
Batas atas = Q3 + 1.5 × IQR

Data yang nilainya berada di bawah batas bawah atau di atas batas atas akan dianggap sebagai outlier.

Berdasarkan hasil output, beberapa fitur memiliki outlier. Oleh karena itu, pada tahap berikutnya outlier dapat ditangani menggunakan metode capping, yaitu membatasi nilai ekstrem agar tidak melewati batas bawah dan batas atas yang telah ditentukan.

**PENANGANAN OUTLIER**

Outlier ditangani menggunakan metode IQR capping. Teknik capping tidak menghapus data, tetapi membatasi nilai ekstrem agar tidak melewati batas bawah dan batas atas.

Dengan cara ini, jumlah data tetap dipertahankan, tetapi pengaruh nilai ekstrem dapat dikurangi.

````python
def buat_batas_iqr(data, columns):
    batas = {}

    for col in columns:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        iqr = q3 - q1

        batas_bawah = q1 - 1.5 * iqr
        batas_atas = q3 + 1.5 * iqr

        batas[col] = {
            'Batas_Bawah': batas_bawah,
            'Batas_Atas': batas_atas
        }

    return pd.DataFrame(batas).T
````

````python
def terapkan_capping(data, batas_iqr):
    data_capped = data.copy()

    for col in batas_iqr.index:
        batas_bawah = batas_iqr.loc[col, 'Batas_Bawah']
        batas_atas = batas_iqr.loc[col, 'Batas_Atas']

        data_capped[col] = data_capped[col].clip(
            lower=batas_bawah,
            upper=batas_atas
        )

    return data_capped
````

Kode tersebut digunakan untuk membuat fungsi penanganan outlier menggunakan metode IQR capping. Pada bagian pertama, fungsi buat_batas_iqr() digunakan untuk menghitung batas bawah dan batas atas pada setiap fitur numerik.

Di dalam fungsi tersebut, setiap kolom dihitung nilai Q1, Q3, dan IQR. Setelah itu, batas bawah dan batas atas ditentukan menggunakan rumus:

Batas bawah = Q1 - 1.5 × IQR
Batas atas = Q3 + 1.5 × IQR

Hasil dari fungsi ini berupa tabel yang berisi Batas_Bawah dan Batas_Atas untuk setiap fitur.

Selanjutnya, fungsi terapkan_capping() digunakan untuk menerapkan batas tersebut pada data. Jika terdapat nilai yang lebih kecil dari batas bawah, maka nilai tersebut akan diganti menjadi batas bawah. Jika terdapat nilai yang lebih besar dari batas atas, maka nilai tersebut akan diganti menjadi batas atas.

Dengan kata lain, outlier tidak dihapus, tetapi nilainya dibatasi agar tidak terlalu ekstrem. Teknik ini dipilih karena jumlah data tetap dipertahankan, tetapi pengaruh nilai ekstrem terhadap proses modelling dapat dikurangi.

Dengan demikian, evaluasi model tetap dilakukan pada data validasi asli sehingga dapat mengurangi risiko data leakage.


**MELATIH DAN MENGEVALUASI SEMUA MODEL**

````python
results = []

for model_name, pipeline in pipelines.items():
    print("Melatih model:", model_name)

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_val)

    accuracy = accuracy_score(y_val, y_pred)

    precision_macro = precision_score(
        y_val,
        y_pred,
        average='macro',
        zero_division=0
    )

    recall_macro = recall_score(
        y_val,
        y_pred,
        average='macro',
        zero_division=0
    )

    f1_macro = f1_score(
        y_val,
        y_pred,
        average='macro',
        zero_division=0
    )

    results.append({
        'Model': model_name,
        'Accuracy': accuracy,
        'Precision_macro': precision_macro,
        'Recall_macro': recall_macro,
        'F1_macro': f1_macro
    })

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(
    by=['Accuracy', 'F1_macro'],
    ascending=False
).reset_index(drop=True)

results_df
````

Kode tersebut digunakan untuk **melatih dan mengevaluasi seluruh model klasifikasi** yang sebelumnya sudah dimasukkan ke dalam pipeline.

Pertama, dibuat list kosong bernama `results` untuk menyimpan hasil evaluasi dari setiap model. Kemudian dilakukan perulangan pada setiap pipeline menggunakan:

```python
for model_name, pipeline in pipelines.items():
```

Pada setiap perulangan, model dilatih menggunakan data training:

```python
pipeline.fit(X_train, y_train)
```

Setelah model selesai dilatih, model digunakan untuk memprediksi data validasi:

```python
y_pred = pipeline.predict(X_val)
```

Hasil prediksi tersebut kemudian dibandingkan dengan nilai asli pada `y_val` untuk menghitung beberapa metrik evaluasi, yaitu **accuracy**, **precision macro**, **recall macro**, dan **f1 macro**.

`accuracy` digunakan untuk melihat persentase prediksi yang benar secara keseluruhan. `precision_macro` digunakan untuk melihat rata-rata ketepatan prediksi pada seluruh kelas. `recall_macro` digunakan untuk melihat rata-rata kemampuan model dalam mengenali setiap kelas. `f1_macro` digunakan untuk melihat keseimbangan antara precision dan recall pada seluruh kelas.

Penggunaan metrik macro penting karena distribusi target `quality` tidak seimbang. Dengan macro average, setiap kelas mendapatkan perhatian yang sama dalam evaluasi, sehingga performa model tidak hanya dinilai dari kelas mayoritas seperti quality 5 dan 6.

Setelah setiap model dievaluasi, hasilnya disimpan ke dalam list `results`, kemudian diubah menjadi dataframe `results_df`. Dataframe tersebut diurutkan berdasarkan nilai `Accuracy` dan `F1_macro` tertinggi.
